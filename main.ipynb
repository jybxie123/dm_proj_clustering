```python
import clustering as cl
```

```python
clustering_methods = [ 'dbscan' ] # 'kmeans', 'kmeans_plus_plus',
datasets = ['iris', 'wine', 'digits']
y_num = {'iris': 3, 'wine': 3, 'digits': 10}
eps = {'iris': 0.4, 'wine': 19, 'digits': 0.0001}
min_samples = {'iris': 5, 'wine': 14, 'digits': 3}
for m in clustering_methods:
    for d in datasets:
        c = cl.Clustering(m, d, y_num[d], eps[d], min_samples[d])
        c.train()
```
