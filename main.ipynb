```python
import clustering as cl
```

```python
clustering_methods = ['kmeans', 'kmeans_plus_plus'] # 'dbscan' 
datasets = ['iris', 'mall_cust', 'blobs']
y_num = {'iris':3, 'mall_cust': 5, 'blobs':4}
for m in clustering_methods:
    for d in datasets:
        c = cl.Clustering(m, d, y_num[d])
        c.train()
```
