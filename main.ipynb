```python
clustering_methods = ['kmeans', 'kmeans_plus_plus'] # 'dbscan' 
datasets = ['iris', 'mall_cust', 'blobs', 'blobs_with_outlier', 'engy_time', 'two_diamonds']
y_num = {'iris':3, 'mall_cust': 5, 'blobs':3, 'blobs_with_outlier':3, 'engy_time':2, 'two_diamonds':2}
for m in clustering_methods:
    for d in datasets:
        c = cl.Clustering(m, d, y_num[d])
        c.train()
```
