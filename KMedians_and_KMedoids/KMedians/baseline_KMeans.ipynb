```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets

# 使用肘部法选择最佳聚类数量k

# 加载数据
iris = datasets.load_iris()
X = iris.data

# 计算不同k值的SSE
sse = []
k_range = range(2, 11)  # 通常从1到10测试不同的k值
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)  # 每个模型的inertia属性都是SSE

# 绘制肘部图
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of clusters k')
plt.ylabel('Total sum of squared distances within clusters')
plt.title('# of clusters v.s. total sum of sqr dis (K-Means)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets

# 加载数据
iris = datasets.load_iris()
X = iris.data

# 选择聚类数
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

# 聚类标签和中心
labels = kmeans.labels_
centers = kmeans.cluster_centers_

plt.figure(figsize=(10, 6))

# 获取 'viridis' 色图
cmap = plt.get_cmap('viridis')
# 计算每个簇颜色的索引
colors = [cmap(i/(k-1)) for i in range(3)]  # 确保颜色均匀分布
for i in range(k):  # 遍历三个簇
    # 选择当前簇的数据点
    cluster_points = X[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=colors[i], label=f'Cluster {i+1}')

# plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', label='Clusters')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title('K-Means Clustering on Iris Dataset (1st and 2nd dimensions)')
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets

# 加载数据
iris = datasets.load_iris()
X = iris.data

# 选择聚类数
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# 聚类标签和中心
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# 选择两个维度进行绘图，例如第一个维度和第三个维度（索引为0和2）
plt.figure(figsize=(10, 6))

# 获取 'viridis' 色图
cmap = plt.get_cmap('viridis')
# 计算每个簇颜色的索引
colors = [cmap(i/(k-1)) for i in range(3)]  # 确保颜色均匀分布
for i in range(k):  # 遍历三个簇
    # 选择当前簇的数据点
    cluster_points = X[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 2], c=colors[i], label=f'Cluster {i+1}')

# plt.scatter(X[:, 0], X[:, 2], c=labels, cmap='viridis', marker='o', label='Clusters')
plt.scatter(centers[:, 0], centers[:, 2], c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title('K-Means Clustering on Iris Dataset (1st and 3rd dimensions)')
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import make_blobs

def get_blobs_with_outlier():
    X, y = make_blobs(n_samples=1000, centers=3, cluster_std=0.60, random_state=0)
    np.random.seed(42)  # 确保生成的离群值是可重复的
    n_outliers = 100
    outliers = np.random.uniform(low=-10, high=10, size=(n_outliers, 2))
    outlier_labels = -1 * np.ones((n_outliers,))
    X = np.vstack([X, outliers])
    y = np.concatenate([y, outlier_labels])
    return (X, y)

get_blobs_with_outlier()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

blobs = get_blobs_with_outlier()[0]
blobs_class = get_blobs_with_outlier()[1]

# 设置随机种子以获得可重复结果
np.random.seed(42)
initial_medians_indices = np.random.choice(len(blobs), 3, replace=False)
initial_medians = blobs[initial_medians_indices]

# 初始化KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(blobs)

# 获取聚类标签和中心
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# 绘图
plt.figure(figsize=(10, 6))
cmap = plt.get_cmap('viridis')

# 绘制所有数据点
scatter = plt.scatter(blobs[:, 0], blobs[:, 1], c=labels, cmap='viridis', marker='o')

# 绘制聚类中心点
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')

# 添加图例
# 为每个聚类生成一个颜色条目
for i in range(3):
    plt.scatter([], [], color=cmap(i / 3), label=f'Cluster {i+1}')

plt.title('K-Means Clustering on Blobs')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
